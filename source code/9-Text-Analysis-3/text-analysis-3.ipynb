## 9. 다음 영화 리뷰 탐색하기
- 리뷰 데이터 토큰화
- ```nltk```를 활용한 텍스트 탐색

### 리뷰 데이터 불러오기
- ```open()``` 함수를 통해 파일을 열며, 열때 "r" (읽기) 모드로 설정한다.

````python
file = open("4-4.txt", "r", encoding="utf-8")
data = file.read().split("\n")
file.close()
````

### 리뷰 데이터 토큰화
- ```konlpy```의 ```morphs()``` 함수를 활용해 토큰화한다
- 모든 토큰을 하나의 리스트 내에 저장한다(for문 활용)

````python
from konlpy.tag import Twitter
t = Twitter()
data_tokenized = []

for sentence in data:
    data_tokenized += t.morphs(sentence)
    
print(data_tokenized[:10])
````

### nltk를 활용한 텍스트 탐색
- ```nltk``` 모듈의 ```Text()```를 활용해 텍스트를 탐색한다

````python
from nltk import Text

text = Text(data_tokenized)
````

````python
## 텍스트의 토큰 추출하기(결과: 리스트)
all_tokens = text.tokens                # 모든 토큰
unique_tokens = set(all_tokens)         # 유니크한 토큰

print("모든 토큰의 개수: ", len(all_tokens))
print("유니크한 토큰의 개수: ", len(unique_tokens))
````

````python
## 각 토큰의 출현 횟수 시각화
# 폰트와 figure 및 label의 size 세팅
from matplotlib import font_manager, rc
font_name = font_manager.FontProperties(fname='c:/windows/fonts/malgun.ttf').get_name()
rc('font', family=font_name)
rc("figure", figsize = (50, 30))
rc("xtick", labelsize = 30)

text.plot(50)
````

````python
## 유사한 단어
text.similar("주인공")
text.similar("세바스찬")
text.similar("현실")
````

````python
## 연어(collocations)
text.collocations()
````
