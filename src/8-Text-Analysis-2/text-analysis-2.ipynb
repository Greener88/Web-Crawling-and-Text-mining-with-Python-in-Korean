### 리뷰 데이터 불러오기
- ```open()``` 함수를 통해 파일을 열며, 열때 "r" (읽기) 모드로 설정한다.
- 파일 읽어오기
    - ```read()```
    - ```readlines()```
    - ```readline()```

````python
# read() 함수를 통해 불러오기
file = open("4-3.txt", "r", encoding="utf-8")
c1 = file.read()
file.close()
print(c1)
````

````python
# readlines() 함수를 통해 불러오기
file = open("4-3.txt", "r", encoding="utf-8")
c2 = file.readlines()
file.close()
print(c2)
````

### 줄바꿈 문자(```\n```)을 없애려면?
- ```read()```나 ```readlines()``` 함수 어떤걸 사용하더라도 각 리뷰의 끝에 줄바꿈 문자(```\n```)가 달려있는 것을 볼 수 있다. 이를 없애려면 어떻게 할 수 있을까?
    - ```split()``` 함수 사용(```read()``` 사용시): string을 list로 변환
    - ```replace()``` 함수 사용(```readlines()``` 사용시): list를 list로 변환

````python
file = open("4-3.txt", "r", encoding="utf-8")
c1 = file.read().split("\n")
file.close()
print(c1)
````

````python
file = open("4-3.txt", "r", encoding="utf-8")
c2 = [x.replace("\n", "") for x in file.readlines()]
file.close()
print(c2)
````
